## Pergunta de Negócio 1:
Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

Demonstre o resultado através de um gráfico de barras.

A cidade com maior valor é New York

```python
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/Usuario/OneDrive/Documentos/SI - CAROL/Sistemas de Informação 04/Ciência de Dados/Arquivos/dataset_3.csv')

# Filtrando por suplementos de escritório
categoria_office = df.loc[df['Categoria']=='Office Supplies',['Cidade', 'Categoria', 'Valor_Venda']]
categoria_office
```

```python
#Somando o total de vendas de cada cidade
#Lembrar de colocar o reset_index para continuar os valores das tabelas depois do groupby
total_cidades = categoria_office.groupby('Cidade')['Valor_Venda'].sum().reset_index()

# Pegando os maiores 10 valores porque não seria possível ver todos eles no gráfico
total_cidades = total_cidades.nlargest(10,'Valor_Venda')

total_cidades
```

```python
# Fazendo gráfico
x = total_cidades['Cidade']
y = total_cidades['Valor_Venda']

plt.figure(figsize=(14,5))
plt.bar(x, y, color = 'pink')
plt.xlabel('Cidades')
plt.ylabel('Total de Vendas')
plt.show()
```

```python
# Teste para ver se tá certo
categoria_office = categoria_office.loc[(categoria_office['Cidade']=='Woodstock') & (categoria_office['Categoria']== 'Office Supplies'), ['Cidade', 'Valor_Venda', 'Categoria']]
categoria_office
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
df7 = pd.read_csv('C:/Users/Usuario/OneDrive/Documentos/SI - CAROL/Sistemas de Informação 04/Ciência de Dados/Arquivos/dataset_3.csv')
cot = 0
for i in range(len(df7['Valor_Venda'])):
    if df7['Valor_Venda'][i] > 1000:
        # Usar o at quando quero atribuir um valor a uma determinada posição da tabela
        df7.at[i, 'Valor_Venda'] = df7['Valor_Venda'][i] * (75/100)
        cot = cot + 1
    else:
        df7.at[i, 'Valor_Venda'] = df7['Valor_Venda'][i] * (80/100)
df7
```

```python
print('Receberiam 15% de desconto um total de', cot, 'vendas')
```

## Pergunta de Negócio 9 (Desafio):

Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
df9 = pd.read_csv('C:/Users/Usuario/OneDrive/Documentos/SI - CAROL/Sistemas de Informação 04/Ciência de Dados/Arquivos/dataset_3.csv')
df9['Data_Pedido'] = pd.to_datetime(df['Data_Pedido'], format='%d/%m/%Y')
df9['Ano'] = df9['Data_Pedido'].dt.year
df9['Mes'] = df9['Data_Pedido'].dt.month
```

```python
media_ano = df9.groupby(['Ano'])['Valor_Venda'].mean()
media_mes = df9.groupby(['Mes'])['Valor_Venda'].mean()
media_seg = df9.groupby(['Segmento'])['Valor_Venda'].mean()
```

```python
mes = df9['Mes'].unique()
ano = df9['Ano'].unique()
seg = df9['Segmento'].unique()
seg
```

```python
# Gráfico de Linhas Ano
x_ano = ano
y_ano = media_ano

x_mes = mes
y_mes = media_mes

x_seg = seg
y_seg = media_seg

fig, axes = plt.subplots(1,3, figsize= (20,5))
axes[0].plot(x_ano, y_ano, color = '#FFA500')
axes[0].set_title("Média de Vendas por Ano")
axes[0].set_xlabel("Ano")
axes[0].set_ylabel("Média de Vendas")

axes[1].plot(x_mes, y_mes, color='#9B59B6')
axes[1].set_title("Média de Vendas por Mes")
axes[1].set_xlabel("Mes")
axes[1].set_ylabel("Média de Vendas")

axes[2].plot(x_seg, y_seg, color='#006400')
axes[2].set_title("Média de Vendas por Segmento")
axes[2].set_xlabel("Segmento")
axes[2].set_ylabel("Média de Vendas")
plt.show()
```

## Pergunta de Negócio 10 (Desafio):

Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias?

Demonstre tudo através de um único gráfico.

```python
sub_total = df.groupby(['SubCategoria']).sum().reset_index()
top12 = sub_total.nlargest(12, 'Valor_Venda')
sub_soma = top12.groupby(['SubCategoria'])['Valor_Venda'].sum().reset_index()
cat_lista = df[df['SubCategoria'].isin(top12['SubCategoria'])]
cat_soma = cat_lista.groupby(['Categoria'])['Valor_Venda'].sum().reset_index()
```

```python
x1 = cat_soma['Categoria']
y1 = cat_soma['Valor_Venda']

x2 = sub_soma['SubCategoria'] 
y2 = sub_soma['Valor_Venda']

plt.figure(figsize =(20,7))
plt.bar(x1, y1, label = 'Categoria', color = '#ADD8E6')
plt.bar(x2, y2, label = 'SubCategoria', color='pink')
plt.legend()
plt.show()
```
